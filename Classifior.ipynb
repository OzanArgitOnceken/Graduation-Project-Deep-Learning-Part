```python
import os 
import shutil
from PIL import Image
```

```python
def extract_values(entry):
    parts = entry.replace(',', '.').split("X") 
    value1 = float(parts[0])
    value2 = float(parts[1]) 
    return value1, value2
```

```python
def sortByXY(directory, x, y):    # Dizin içindeki dosya isimlerini al
    file_names = os.listdir(directory)
    
    # Dosya isimlerini sırala 2. sütuna göre
    sorted_file_names = sorted(file_names, key=lambda file_name: extract_values(file_name)[1])
    values = []
    # Sadece 2. sütun değerlerini ekrana yazdır
    print(f"Sıralı Dosya İsimleri (Value{x} veya Value{y}'ye göre 2. Sütun Değerleri):")
    for file_name in sorted_file_names:
        value2 = extract_values(file_name)[1]
        values.append(value2)
        print(value2)
    ave=sum(values) / len(values)
    print()
    print(ave)
```

```python
def move_files(source_folder, target_folder):
    # Get all file names in the source folder
    file_names = os.listdir(source_folder) 
    for file_name in file_names:
        file_path = os.path.join(source_folder, file_name) 
        xValue,yValue=extract_values(file_name)
        specifiedFileName=""
        
        if xValue>=0.02:
            specifiedFileName+="Right"
        else:
            specifiedFileName+="Left"
        
        if yValue>=0.23:
            specifiedFileName+="Up"
        else:
            specifiedFileName+="Down"
        
        subdirectory_path = os.path.join(target_folder, specifiedFileName)     
        
        target_path = os.path.join(subdirectory_path, file_name)

        shutil.move(file_path, target_path)
            
        print(xValue,yValue,specifiedFileName)
 
```

```python
source_folder = "source"
target_folder = "target"
move_files(source_folder,target_folder)
#move_files(source_folder, target_folder)
```
